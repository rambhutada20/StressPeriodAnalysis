```python
import yfinance as yf
import pandas as pd
import numpy as np
```

```python
SPY  = yf.download("SPY" , start= "2005-01-01", end = "2026-03-25")
QQQ  = yf.download("QQQ" , start= "2005-01-01", end = "2026-03-25")
XLE  = yf.download("XLE" , start= "2005-01-01", end = "2026-03-25")
GLD = yf.download("GLD" , start= "2005-01-01", end = "2026-03-25")
TLT  = yf.download("TLT" , start= "2005-01-01", end = "2026-03-25")
```

```python
SPY
```

```python
SPY.columns
```

```python
column =  [ (  'High', 'SPY'),
            (   'Low', 'SPY'),
            (  'Open', 'SPY'),
            ('Volume', 'SPY')] 
```

```python
SPY = SPY.drop(columns= column)
```

```python
SPY["Return"] = np.log(SPY[( 'Close', 'SPY')] / SPY[( 'Close', 'SPY')].shift(1))
```

```python
SPY
```

```python
SPY = SPY.rename(columns = {'Return': 'SPY_Return'})
```

```python
SPY = SPY.drop(columns = [( 'Close', 'SPY')])
```
